```python
cols = []
cols_values = [[] for i in range(13)]
for i in range(len(rows)):
    if i == 0:
        cols = rows[0].find_all("th")
        cols = [str(col).replace("<th>", "").replace("</th>", "") for col in cols]
        print(len(cols))
    else:
        row = rows[i].find_all("td")
        row = [str(r).replace("<td>", "").replace("</td>", "") for r in row]
#         print(row)
        for i in range(len(row)):
            cols_values[i].append(row[i])
            
final_dict = {}
for i in range(13):
    final_dict[cols[i]] = cols_values[i]
final_dict["System_ID"] = [system] * (len(rows) - 1)
final_dict["Community"] = [community] * (len(rows) - 1)
```

### Methods

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def extract_legal_names():
    url = "https://licensing.copyright.gov/lds/servlet/lds.SearchLegalNameCTL"
    payload = { "selectedLegalName" : "%" }
    response = requests.post(url, data=payload)
    soup = BeautifulSoup(response.text)
    select_tag = soup.find_all("select", { "name" : "communities" })
    communities = select_tag[0].find_all("option")
    communities = [community.get("value") for community in communities]
    return communities

def extract_system_ids(legal_name):
    url = "https://licensing.copyright.gov/lds/servlet/lds.DisplayLegalNameCTL"
    data = { "communities" : legal_name}
    response = requests.post(url, data=data)   
    soup = BeautifulSoup(response.text)
    select_tag = soup.find_all("select", { "name" : "communities"})
    communities = select_tag[0].find_all("option")
    if len(communities) >= 1:
        system_ids = [community.get("value").split(" ")[0] for community in communities]
        return system_ids
    
    return []
    

def extract_statement_of_accounts(system_id):
    data = { "selectedID" : system_id }
    url = "https://licensing.copyright.gov/lds/servlet/lds.SearchIDCTL"
    response = requests.post(url, data=data)
    soup = BeautifulSoup(response.text)
    table = soup.find_all("table", { "class" : "compact display"})
    rows = table[0].find_all("tr")
    if len(rows) > 1:
        main_text = soup.find_all("div", {"class" : "main_text"})
        h4s = main_text[0].find_all("h4")
        system = str(h4s[0]).replace("<h4 align=\"left\">System ID:", "").replace("</h4>", "").strip()
        community = str(h4s[1]).replace("<h4 align=\"left\">Community:", "").replace("</h4>", "").strip()
        cols = []
        cols_values = [[] for i in range(13)]
        for i in range(len(rows)):
            if i == 0:
                cols = rows[0].find_all("th")
                cols = [str(col).replace("<th>", "").replace("</th>", "") for col in cols]
                print(len(cols))
            else:
                row = rows[i].find_all("td")
                print(row)
                row = [str(r).replace("<td>", "").replace("</td>", "") for r in row]
        #         print(row)
                for i in range(len(row)):
#                     print(row[i])
                    cols_values[i].append(str(row[i]))

        final_dict = {}
        for i in range(13):
            final_dict[cols[i]] = cols_values[i]
        final_dict["System_ID"] = [system] * (len(rows) - 1)
        final_dict["Community"] = [community] * (len(rows) - 1)
        df = pd.DataFrame(final_dict)
        return df
#         df.to_csv("library_congress.csv", mode="a", index=True, header=False)
```
